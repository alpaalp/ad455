```r
billboard %>% filter(wk1> 80 & wk2 >65)
```

```r
billboard %<>% mutate(starting_performance_point = wk1/10)
```

```r
billboard
```

```r
billboard %>% mutate(performance_comment = ifelse(starting_performance_point >8,"excellent",ifelse(starting_performance_point <6,"vasatin_alti","orta")))
```

```r
billboard %>%
group_by(track) %>%
summarise(max_on_top_3weeks = max(wk1,wk2,wk3, na.rm = F))
```

```r
billboard_ex <- billboard %>%
group_by(track) %>%
summarise(max_track_time_3weeks = max(wk1, na.rm = T),
          min_track_time_3weeks = min(wk1, na.rm = T),
          av_track_time_3weeks = mean(wk1, na.rm = T))
```
